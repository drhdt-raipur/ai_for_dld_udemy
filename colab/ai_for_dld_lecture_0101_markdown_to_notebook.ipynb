from IPython.display import HTML, display
import nbformat as nbf
import uuid
from google.colab import files

# Step 1: Show HTML input UI
html_ui = """
<textarea id="Input" style="width:100%;height:300px;" placeholder="Paste your  here..."></textarea><br>
<button onclick="google.colab.kernel.invokeFunction('notebook.convert__to_ipynb', [document.getElementById('Input').value], {})">
Generate Notebook and Download
</button>
"""
display(HTML(html_ui))

# Step 2: Define conversion function
def convert__to_ipynb(_text):
    lines = _text.split('\n')
    cells = []
    current_cell_lines = []
    cell_type = ''

    for line in lines:
        if line.strip().startswith(""):
            if current_cell_lines:
                cells.append(
                    nbf.v4.new__cell('\n'.join(current_cell_lines)) if cell_type == ''
                    else nbf.v4.new_code_cell('\n'.join(current_cell_lines))
                )
                current_cell_lines = []
            cell_type = 'code' if cell_type == '' else ''
        else:
            current_cell_lines.append(line)

    if current_cell_lines:
        cells.append(
            nbf.v4.new__cell('\n'.join(current_cell_lines)) if cell_type == ''
            else nbf.v4.new_code_cell('\n'.join(current_cell_lines))
        )

    # Create notebook
    nb = nbf.v4.new_notebook()
    nb['cells'] = cells
    nb['metadata'] = {
        "kernelspec": {
            "display_name": "Python 3",
            "language": "python",
            "name": "python3"
        },
        "language_info": {
            "name": "python",
            "version": "3.10"
        }
    }

    filename = f"generated_{uuid.uuid4().hex[:6]}.ipynb"
    with open(filename, 'w', encoding='utf-8') as f:
        nbf.write(nb, f)

    files.download(filename)

# Step 3: Register callback
from google.colab import output
output.register_callback('notebook.convert__to_ipynb', convert__to_ipynb)
